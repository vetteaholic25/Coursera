## Reading the data into a dataframe and display the top 5 rows

```python
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/Topic_Survey_Assignment.csv'
df_interests = pd.read_csv(url)
df_interests.head()
```

## Processing the data and making a bar chart

```python
df_interests.sort_values('Very interested', ascending = False, inplace = True)
#(df_interests['Very interested'] / df_interests['Very interested'].sum(axis=1)) * 100
df_interests = df_interests.div(df_interests.sum(axis=1), axis=0)*100
#((df_interests / df_interests.sum(axis=1))* 100).round(1).astype(str) + '%'
```

```python
colors_list = ['#5cb85c','#5bc0de','#d9534f']
ax = df_interests.plot(kind = 'bar', figsize = (20, 8), width = 0.8, color = colors_list, edgecolor = None)
plt.title("Percentage of Respondents' Interest in Data Science Areas",fontsize= 16)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
```
